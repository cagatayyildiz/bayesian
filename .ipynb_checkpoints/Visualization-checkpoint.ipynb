## 2-D Plots

For the complete list of  examples, see http://www.labri.fr/perso/nrougier/teaching/matplotlib/

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show()
```
